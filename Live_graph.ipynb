```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
import time
import psutil
#third party library which gives out the cpu usage
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
#for configuring matplotlib so that we can work on live graph
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
#single subplot on the first row and the first column
fig.show()
```

```python
psutil.cpu_percent()
#gives out your current cpu usage
```

```python
#for real time data
i = 0
x, y = [], [] #two empty lists

while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    
    ax.plot(x, y, color='b')
    
    fig.canvas.draw()
    
    time.sleep(0.1) #for delay
    i +=1 #for increasing iteration value/counter value

```
