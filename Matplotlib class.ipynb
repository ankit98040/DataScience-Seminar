```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)        
y = x ** 2
```

```python
x
```

```python
y
```

```python
#plt.plot(x,y,'g')     # rgb format is directly supported

plt.plot(x,y,color = 'orange')
```

```python
plt.plot(x,y,color = 'orange')

plt.xlabel('X axis label here ')
plt.ylabel('Y axis label here')
plt.title('Title here ')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r*-') 
plt.subplot(1,2,2)
plt.plot(y, x, 'g+-')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9]) 
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.5])

axes1.plot(x,y,'r')

axes2.plot(y,x,'g')
```

```python
fig.savefig("filename.png", dpi = 600)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="hello")
ax.plot(x, x**3, label="x**3")
ax.legend(loc = 0)  #0-10
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b*-') # blue line with dots

ax.plot(x, x**3, 'g', lw=10, alpha=0.5, ls='--', marker='o', markersize=10, 
        markerfacecolor='yellow', markeredgewidth='5', markeredgecolor='red')
#alpha is used for transparency of the line
#ls = linestyle     lw = line width
```

```python
plt.scatter(data['TotalPay'],data['BasePay'],color = 'red')
```

```python
plt.scatter(data['TotalPay'],data['BasePay'],color = 'red',label = 'dots')
plt.plot(data['TotalPay'],data['BasePay'], label = 'line')
plt.legend()
```

```python
fig=plt.figure()

g1= fig.add_subplot(1,2,1)
g1.plot(data['BasePay'],data['TotalPay'])
g1.scatter(data['BasePay'],data['TotalPay'],color='pink')

g2= fig.add_subplot(1,2,2)
g2.plot(data['OtherPay'],data['TotalPay'])
g2.scatter(data['OtherPay'],data['TotalPay'],color='pink')

plt.tight_layout()
```

```python
fig=plt.figure()

fig.patch.set_facecolor('pink')

g1= fig.add_subplot(1,2,1, facecolor='blue')
g1.plot(data['BasePay'],data['TotalPay'])
g1.scatter(data['BasePay'],data['TotalPay'],color='pink')

g2= fig.add_subplot(1,2,2, facecolor='green')
g2.plot(data['OtherPay'],data['TotalPay'])
g2.scatter(data['OtherPay'],data['TotalPay'],color='pink')

plt.tight_layout()
```
