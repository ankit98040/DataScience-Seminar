```python
# lambda function - Row wise operation
def new_duration(network, duration):
    if network=='world':
        new_duration = duration * 2
    else:
        new_duration = duration * 4
    
    return new_duration

df['new_duration'] = df.apply(lambda x: new_duration(x['network'], x['duration']), axis=1)
```
